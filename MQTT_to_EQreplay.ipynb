```python
def MQTT_to_EQreplay(file_name):
    file_path = os.path.join(os.getcwd(), file_name)
    try:
        with open(file_path) as mqtt_file:
            try:
                data = mqtt_file.readlines()
                print("取得檔案成功")
                return data
            except:
                return "資料格式錯誤"
    except:
        return "找不到mqtt檔案"
    
```
